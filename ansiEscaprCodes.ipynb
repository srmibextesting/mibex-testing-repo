# ANSI Escape Code Examples in Jupyter

### Example 1: Simple Colored Text (May not work in Jupyter UI)
These examples may not render properly in the standard Jupyter Notebook interface, but should work in a terminal environment like JupyterLab terminal.

```python
print("\x1b[31mThis text is red!\x1b[0m")
print("\x1b[32mThis text is green!\x1b[0m")
print("\x1b[33mThis text is yellow!\x1b[0m")
print("\x1b[34mThis text is blue!\x1b[0m")
```

### Example 2: Combining Colors and Styles

```python
print("\x1b[1m\x1b[31mBold and red text\x1b[0m")
print("\x1b[4m\x1b[32mUnderlined and green text\x1b[0m")
```

### Example 3: Background Colors

```python
print("\x1b[41mThis text has a red background!\x1b[0m")
print("\x1b[42mThis text has a green background!\x1b[0m")
print("\x1b[44mThis text has a blue background!\x1b[0m")
```
